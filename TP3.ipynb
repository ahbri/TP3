```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
```

L'importation des bibliothèques qu'on va travailler avec.

```python
X, y = make_regression(n_samples=100, n_features=1, noise=0.1)
```

Créons ensuite le dataset en utilisant make_regression pour générer un
problème de régression aléatoire de 100 exemples avec une seule variable.


1.    X est la variable indépendante (features) et y est la variable dépendante (target)
2.   n_samples définit le nombre d'exemples à générer

3.   n_features définit le nombre de caractéristiques (variables indépendantes)
4.   noise définit le niveau de bruit ajouté aux données





```python
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Visualisation des données')
plt.show()
```

 Visualisation des données

```python
model = SGDRegressor(max_iter=100, alpha=0.0001)
```

max_iter définit le nombre maximal d'itérations pour la convergence du modèle
et alpha est le terme de régularisation

```python
#entrainer le modèle
model.fit(X, y)
```

```python
#Calculons la précision du modèle
accuracy = model.score(X, y)
print("Coefficient de détermination :", accuracy)
```

Score renvoie le coefficient de détermination R^2 de la prédiction par rapport aux vraies valeurs.


```python
# Faire de nouvelles prédictions et tracer les résultats
predictions = model.predict(X)
# Prédit les valeurs de y pour les données X
plt.scatter(X, y)  # Trace le nuage de points des données originales
plt.plot(X, predictions, color='red')  # Trace la ligne de prédiction en rouge
plt.xlabel('X')  # Ajoute une étiquette à l'axe des x
plt.ylabel('Y')  # Ajoute une étiquette à l'axe des y
plt.title('Prédictions du modèle')  # Ajoute un titre au graphique
plt.show()  # Affiche le graphique
```

```python
#Referons le meme travaille mais cette fois ci avec un nombre d'itérations égale à 1000 et le learning rate égale à 0.001
model_1000 = SGDRegressor(max_iter=1000, alpha=0.001)
model_1000.fit(X, y)
score_1000 = model_1000.score(X, y)
print("Coefficient de détermination pour 1000 itérations :", score_1000)
predictions_1000 = model_1000.predict(X)
# Prédit les valeurs de y pour les données X
plt.scatter(X, y)  # Trace le nuage de points des données originales
plt.plot(X, predictions_1000, color='red')  # Trace la ligne de prédiction en rouge
plt.xlabel('X')  # Ajoute une étiquette à l'axe des x
plt.ylabel('Y')  # Ajoute une étiquette à l'axe des y
plt.title('Prédictions du modèle de 1000 itérations')  # Ajoute un titre au graphique
plt.show()  # Affiche le graphique
```

En examinant les résultats de notre modèle de régression linéaire avec SGDRegressor, nous pouvons constater quelques observations importantes. Tout d'abord, en regardant le coefficient de détermination , nous remarquons que pour la configuration avec 100 itérations et un learning rate de 0.0001, le R² est d'environ 0.9999887860162224, ce qui indique que le modèle explique bien une grande partie de la variance des données. Cependant, en augmentant le nombre d'itérations à 1000 et le learning rate à 0.001, nous voyons une légère amélioration avec un coefficient de détermination d'environ 0.9999866882968496, suggérant une meilleure adéquation du modèle aux données avec une convergence légèrement plus élevée.
